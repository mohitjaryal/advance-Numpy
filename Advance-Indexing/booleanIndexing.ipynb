# Boolean indexing 

```python
import numpy as np
```

```python
# creating an array

arr = np.random.randint(1,100,24).reshape(6,4)
print(arr)
```

## when we have to get data by implimenting logic eg. get all even numbers between 1-100

```python
# boolean indexing
# ques 1 - find all numbers greater than 50

print(arr > 50) 
```

```python
# find all numbers greater than 50 and even
print((arr > 50 ) & (arr % 2 ==0))
```

```python
# find all numbers that are not divisible by 7
print(arr % 7 != 0)
```
