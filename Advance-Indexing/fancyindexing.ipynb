# Fancy indexing 

```python
import numpy as np
```

```python
# creating an array

arr = np.arange(12).reshape(4,3)
print(arr)
```

## suppose we want only 0,2,3 column only then with normal slicing we can't do this.
## so we use fancy indexing in square brackets [] we pass list inside [[]] with column index eg. [[0,2,3]]

```python
# fancy indexing
arr[[0,2,3]]
```
