# Working with missing values 

```python
import numpy as np
```

```python
arr = np.array([0,1,2,3,np.nan,6,7,np.nan,9,np.nan])
```

```python
# checking for the missing values
np.isnan(arr)
```

```python
# now if we wanna remove all the missing values then we'll use bollean indexing
arr[~np.isnan(arr)]
```
