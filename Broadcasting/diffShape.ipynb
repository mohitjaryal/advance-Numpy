# Broadcasting - different shaped array

```python
import numpy as np
```

### Practice ques1

```python
arr1 = np.arange(12).reshape(4,3)
arr2 = np.arange(3)

print(arr1)
print(arr2)

# it will be automatically brodcasted 
print(arr1 + arr2)
```

### Practice ques 2

```python
arr1 = np.arange(3).reshape(1,3)
arr2 = np.arange(3).reshape(3,1)

print(arr1)
print(arr2)

# it will be automatically brodcasted 
print(arr1 + arr2)
```
