# Numpy array vs python lists

## Speed test of Python lists

```python
import time as time
```

```python
# Numpy array vs Python lists

# python's lists
list1 = [i for i in range(10000000)]
list2 = [i for i in range(10000000,20000000)]

list3 = []
```

```python

# starting time
start = time.time()
for i in range(len(list1)):
    list3.append(list1[i] + list2[i])
print(time.time() - start)
```
