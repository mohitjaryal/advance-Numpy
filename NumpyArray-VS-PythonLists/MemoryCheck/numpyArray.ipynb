# Numpy array vs Python lists

## Memory check of numpy array

```python
# importing system module to know how much size python lists occupy
import sys as sys
```

```python
# python's lists
list1 = [i for i in range(10000000)]

# .getsizeof() tells us about the size of the data in bytes (we can acces this by importing sys module)
print(sys.getsizeof(list1))
```
