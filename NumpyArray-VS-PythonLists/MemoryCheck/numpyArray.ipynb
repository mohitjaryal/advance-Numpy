# Numpy array vs Python lists

## Memory check of numpy array

```python
# importing system module to know how much size numpy array occupy
import sys as sys
import numpy as np
```

```python
# numpy array
arr = np.arange(10000000)

# .getsizeof() tells us about the size of the data in bytes (we can acces this by importing sys module)
print(sys.getsizeof(arr))
```
